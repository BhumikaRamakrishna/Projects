```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data_features = data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 5, n_init='auto')
model
```

```python
model.fit(data_features)
```

```python
model.predict(data_features)
```

```python
centres = model.cluster_centers_
centres
```

```python
labels = model.labels_
labels
```

```python
data.columns
```

```python
sns.scatterplot(data, x = 'Annual Income (k$)', y = 'Spending Score (1-100)', s = 50, hue = labels, palette = 'Pastel1' )
plt.scatter(centres[:,0],centres[:, 1], s=200, c='gray', marker = '*')
```
